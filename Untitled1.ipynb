```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor

# 1. Process Hotel Data
print("Cleaning Hotel Data...")
hotel_df = pd.read_csv('hotel_data/hotel_bookings.csv')
hotel_df = hotel_df[hotel_df['is_canceled'] == 0] # Only look at completed trips
hotel_features = hotel_df[['hotel', 'arrival_date_month', 'adults', 'children', 'adr']].copy()
hotel_features = hotel_features.dropna()

# 2. Process Airbnb Data (for US Cities)
print("Cleaning Airbnb Data...")
airbnb_df = pd.read_csv('airbnb_data/train.csv')
airbnb_df['price'] = np.exp(airbnb_df['log_price']) # Convert log to actual dollars
airbnb_features = airbnb_df[['city', 'room_type', 'accommodates', 'price']].copy()

print("Data is now clean and ready!")
```

```python
 # Create the Encoder to turn text into numbers
le = LabelEncoder()
hotel_data_encoded = hotel_features.copy()

# Encode the text columns
for col in ['hotel', 'arrival_date_month']:
    hotel_data_encoded[col] = le.fit_transform(hotel_data_encoded[col].astype(str))

# Define X (Input) and y (Target Price)
X = hotel_data_encoded.drop('adr', axis=1)
y = hotel_data_encoded['adr']

# Train the Random Forest Model
print("Training the AI (this may take 30 seconds)...")
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)
print("AI Training Complete!")
```

```python
def calculate_my_trip(place_type, month, days, adults, children, reason):
    # Prepare input for the model
    # Note: 0 is usually City, 1 is Resort depending on encoding
    p_type = 0 if place_type.lower() == 'city' else 1
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
              'August', 'September', 'October', 'November', 'December']
    m_idx = months.index(month)

    # Predict price per night
    price_per_night = model.predict([[p_type, m_idx, adults, children]])[0]

    # Apply "Reason" Multiplier
    multiplier = 1.0
    if reason.lower() == 'business': multiplier = 1.4
    if reason.lower() == 'study': multiplier = 0.8

    hotel_total = (price_per_night * days) * multiplier
    food_total = (50 * days * (adults + children)) # Suggesting $50 per person per day

    print(f"---YOUR TRAVEL BUDGET ---")
    print(f"Destination: {place_type} | Duration: {days} Days | For: {reason}")
    print(f"Predicted Hotel Cost: ${hotel_total:.2f}")
    print(f"Predicted Food/Living: ${food_total:.2f}")
    print(f"-----------------------------")
    print(f"TOTAL ESTIMATED BUDGET: ${hotel_total + food_total:.2f}")

# TEST IT HERE:
calculate_my_trip(place_type='Resort', month='July', days=7, adults=2, children=1, reason='Leisure')
```

```python
def calculate_total_trip_pro(origin, destination, month, days, adults, reason):
    # 1. Get Hotel Prediction (from our ML Model)
    p_type = 0 if 'city' in destination.lower() else 1
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
              'August', 'September', 'October', 'November', 'December']
    m_idx = months.index(month)
    hotel_price_night = model.predict([[p_type, m_idx, adults, 0]])[0]

    # 2. Flight Logic (Simplified Distance Logic)
    # This is a 'heuristic'  an AI rule of thumb
    flight_cost = 500 # Base price for international
    if origin.lower() == destination.lower(): flight_cost = 100

    # 3. Final Calculation
    hotel_total = hotel_price_night * days
    food_total = 50 * days * adults
    grand_total = hotel_total + food_total + (flight_cost * adults)

    print(f"--- COMPLETE TRAVEL ITINERARY ---")
    print(f"Route: {origin} -> {destination}")
    print(f"Stay: {days} days | Reason: {reason}")
    print(f"------------------------------------")
    print(f"Hotel:   ${hotel_total:.2f}")
    print(f"Flights: ${flight_cost * adults:.2f}")
    print(f"Food/Etc: ${food_total:.2f}")
    print(f"====================================")
    print(f"TOTAL BUDGET: ${grand_total:.2f}")

# Try it!
calculate_total_trip_pro('New York', 'London City', 'August', 7, 2, 'Leisure')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_business_vs_leisure(destination_type='Resort'):
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
              'August', 'September', 'October', 'November', 'December']

    business_costs = []
    leisure_costs = []

    # Simulate a 7-day trip for 2 adults for every month
    for m in months:
        # Get base prediction from ML model
        m_idx = months.index(m)
        p_type = 0 if destination_type.lower() == 'city' else 1
        base_price = model.predict([[p_type, m_idx, 2, 0]])[0] * 7

        # Apply Reason logic: Business is 40% more, Leisure is baseline
        business_costs.append(base_price * 1.4 + (50 * 7 * 2)) # Hotel + Food
        leisure_costs.append(base_price * 1.0 + (50 * 7 * 2))  # Hotel + Food

    # Create the Plot
    plt.figure(figsize=(12, 6))
    plt.plot(months, business_costs, label='Business Trip (Premium)', marker='o', color='red', linewidth=2)
    plt.plot(months, leisure_costs, label='Leisure Trip (Standard)', marker='s', color='blue', linewidth=2)

    plt.fill_between(months, leisure_costs, business_costs, color='gray', alpha=0.2, label='Price Gap')

    plt.title(f'Cost Comparison: Business vs Leisure (7-Day {destination_type} Trip)', fontsize=15)
    plt.ylabel('Total Estimated Budget ($)', fontsize=12)
    plt.xlabel('Month of Year', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.show()

    # Calculate average difference
    avg_diff = sum(business_costs) / 12 - sum(leisure_costs) / 12
    print(f"On average, a Business trip to a {destination_type} costs ${avg_diff:.2f} more than a Leisure trip.")

# Run the comparison
plot_business_vs_leisure('Resort')
```
